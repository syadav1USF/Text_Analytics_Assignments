```python
# Read tweets posted on your own Twitter timeline using REST API
# Tweepy API method home_timeline() retrieves the last 20 entries in your timeline

for tweet in api.home_timeline():
    print(f'{tweet.user.name} said {tweet.text}')
```

```python
# Twitter API endpoints use pagination to return their results. Each method returns 
# the first page only. Tweepy cursors overcomes this problem not only the first page,
# but also from subsequent pages. Cursors are implemented as a Tweepy class requiring 
# two inputs: the API method to fetch results and  number of items wanted.

i = 1
for tweet in tweepy.Cursor(api.home_timeline).items(100):
    print(i, tweet.user.name, tweet.created_at, tweet.text)
    i += 1
```

```python
# List your own tweets
    
for tweet in tweepy.Cursor(api.user_timeline).items():
    print(tweet.created_at, tweet.text)
```

```python
# Search for a string in any public tweet
# q: search string, rpp: how many tweets wanted (max=100), lang='en': English

for tweet in api.search(q='Glassdoor reviews', lang='en', rpp=100):
    print(f'{tweet.user.name} : {tweet.created_at} : {tweet.text}\n')
```

```python
# Save Twitter search results in a Pandas dataframe

import pandas as pd
df = pd.DataFrame(columns=['screen_name', 'name', 'date_time', 'location', 'text'])

for tweet in api.search(q='Glassdoor reviews', lang='en', rpp=100):
    df = df.append(pd.Series([tweet.user.screen_name, tweet.user.name, tweet.created_at, tweet.user.location, tweet.text], index=df.columns), ignore_index=True)

df.shape
```

```python
# Capture streaming tweets using keyword search
# StreamListener receives tweets from the Twitter stream

class MyListener(tweepy.StreamListener):
    def __init__(self, api):
        self.api = api
        self.me = api.me()

    def on_status(self, tweet):
        print(f'{tweet.created_at} : {tweet.user.screen_name} : {tweet.user.location} : \n{tweet.text}\n')
        
    def on_error(self, status):
        print(status)
 
stream = tweepy.Stream(auth, MyListener(api))
stream.filter(track=['Glassdoor reviews'], languages=['en'])
```
